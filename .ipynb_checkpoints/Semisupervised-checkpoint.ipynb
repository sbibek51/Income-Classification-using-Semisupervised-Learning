#Categorical Data Encoding

```python
from sklearn.feature_selection import mutual_info_regression

X = X_resampled[categorical_columns].copy()
y = y_resampled

# Label encoding for categoricals
for colname in X.select_dtypes("object"):
    X[colname], _ = X[colname].factorize()

# All discrete features should now have integer dtypes (double-check this before using MI!)
discrete_features = X.dtypes == int
```

## Mutual Information Computation

```python
def make_mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_regression(X, y, discrete_features=discrete_features)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

mi_scores = make_mi_scores(X, y, discrete_features)
print(f"MI Score:\n\n{mi_scores[::]}")


def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.xlabel('MI Score')
    plt.title("Mutual Information Scores",fontsize=20,fontweight=20)


plt.figure(dpi=100, figsize=(8, 5))
plot_mi_scores(mi_scores)
```

#Supervised Models
***Logistic Regression:*** It's an easy-to-understand method that can handle binary classification tasks and offer a probability score for each prediction.

*Working Principle:* Logistic regression estimates the coefficients of the input characteristics using a mathematical function known as the logistic function. Once the coefficients have been calculated, we can use them to forecast the likelihood of a given input falling into a specific output category.



***Decision Trees:*** It is simple to understand and can handle both category and numerical input characteristics.

*How it works:* Decision trees divide the feature space along the most informative characteristics until each area has just one output category. Following the path down the tree that corresponds to the input characteristics, we may use the tree to predict the output category of a given input.



```python
# Define a function to build preprocessors for numerical and categorical columns
def build_preprocessors(numerical_columns, categorical_columns):
    """Construct preprocessing pipelines for numerical and categorical data."""
    # Define a pipeline for numerical data: handle missing values using mean and scale them
    numerical_transformer = Pipeline([
        ('impute', SimpleImputer(strategy='mean')),
        ('scaling', StandardScaler())
    ])

    # Define a pipeline for categorical data: handle missing values using mode and encode them
    categorical_transformer = Pipeline([
        ('impute', SimpleImputer(strategy='most_frequent')),
        ('encoding', OneHotEncoder(handle_unknown='ignore'))
    ])

    # Combine the numerical and categorical pipelines into a single column transformer
    preprocessor = ColumnTransformer([
        ('num', numerical_transformer, numerical_columns),
        ('cat', categorical_transformer, categorical_columns)
    ])

    return preprocessor
```

```python
# Define a function to train models, predict and print metrics
def model_predictions(models, X_train, y_train, X_test, y_test, preprocessor):
    """Train models and print their metrics."""
    f_scores = []

    for model in models:
        # Define a pipeline to preprocess data and then train the model
        pipeline = Pipeline([
            ('preprocessing', preprocessor),
            ('model', model)
        ])

        # Compute cross validation score for the model
        cross_val_scores = cross_val_score(pipeline, X_train, y_train, cv=5, scoring='accuracy')
        print(f"('Model:') {model}")
        print(f"('Cross Validation score:') {cross_val_scores.mean()}\n")

        pipeline.fit(X_train, y_train)
        # Predict using the model on test data
        pred = pipeline.predict(X_test)

        # Append the f1 score of predictions to the f_scores list
        f_scores.append(f1_score(pred, y_test))

        print_metrics(pred, y_test)

    return f_scores
```

```python
# Define a utility function to print various metrics for the model
def print_metrics(pred, y_test):
    """Utility function to print model metrics."""
    print(f" ('Accuracy Score:') {accuracy_score(pred, y_test)}\n")
    print(f" ('F1 Score:') {f1_score(pred, y_test)}\n")
    print(f" ('Confusion Matrix:') {confusion_matrix(pred, y_test)}")
    print(f" ('Classification Report:') \n{classification_report(pred, y_test)}\n")
```

In the function above, the following libraries are used:

**Pipeline:** This is used to chain multiple data processing steps into a single unit.

**SimpleImputer:** This is used to handle missing values by imputing them with a specified strategy (mean for numerical data, and mode for categorical data).

**StandardScaler:** This is used to scale numerical features to have zero mean and unit variance.

**OneHotEncoder:** This is used to encode categorical features into binary one-hot vectors.


```python
# Define a function to make predictions using semi-supervised models
def semi_supervised_model_predictions(models, X_train, y_train, y_test):
    """Predictions for semi-supervised models."""
    # Define simple imputers for numerical and categorical columns
    si_num = SimpleImputer(strategy='mean')
    si_cat = SimpleImputer(strategy='most_frequent')
    # Define a standard scaler for numerical data
    sc = StandardScaler()

    # Impute and preprocess categorical columns
    X_train_cat = pd.DataFrame(si_cat.fit_transform(X_train[categorical_columns]), columns=categorical_columns)
    # Impute and preprocess numerical columns
    X_train_num = pd.DataFrame(si_num.fit_transform(X_train[numerical_columns]), columns=numerical_columns)

    # Convert categorical columns to dummy variables (one hot encoding)
    X_train_cat = pd.get_dummies(X_train_cat, drop_first=True)
    # Scale the numerical columns
    X_train_num = sc.fit_transform(X_train_num)

    # Concatenate preprocessed categorical and numerical columns
    X_train_final = pd.concat([X_train_cat, pd.DataFrame(X_train_num, columns=numerical_columns)], axis=1)

    scores = []
    # Loop through each model and make predictionsfor model in models:
    for model in models:
        model.fit(X_train_final, y_train)
        pred = model.predict(X_train_final)

        print(f"('Model:') {model}")
        print_metrics(pred, y_test)

        # Append the f1 score of predictions to the scores list
        scores.append(f1_score(y_test, pred, zero_division=1))

    return scores
```

```python
# Instantiate various models
lr = LogisticRegression(max_iter=1000)  # Logistic Regression model with a max iteration of 1000
dt = DecisionTreeClassifier()  # Decision Tree Classifier
svc = svm.SVC(probability=True, gamma='auto')  # Support Vector Classifier with specified parameters
model_svc = semi_supervised.SelfTrainingClassifier(svc)  # Self Training Classifier using SVC as the base model
model_lp = semi_supervised.LabelPropagation()  # Label Propagation model
model_ls = semi_supervised.LabelSpreading()  # Label Spreading model
```

```python
# Build the preprocessor using the defined function
preprocessor = build_preprocessors(numerical_columns, categorical_columns)
```

```python
# Split the resampled data into features and target variables
X = train_resampled[predictors_columns]
y = train_resampled[TARGET]
```

```python
# Split the data into training and testing sets using stratified sampling
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Call the model_predictions function to train the Logistic Regression and Decision Tree models and get their f1 scores
output = model_predictions([lr, dt], X_train, y_train, X_test, y_test, preprocessor)
```

```python
# Create a DataFrame to store model names and their corresponding f1 scores
final = pd.DataFrame({'model': ['LogisticRegression()', 'DecisionTreeClassifier()'], 'F1_score': output})
```

```python
# Define a function to plot model performances based on their f1 scores
def plot_model_performance(df, title):
    """Utility function to plot model performances."""
    plt.figure(figsize=(10, 8))
    sns.barplot(x='model', y='F1_score', data=df) # Barplot of models vs. their f1 scores
    plt.title(title, fontsize=25)
    plt.ylabel('F1-score', fontsize=20)
    plt.xlabel('Models', fontsize=20)
    plt.tight_layout()
    plt.show()
```

```python
# Call the function to plot the f1 scores of the Logistic Regression and Decision Tree models
plot_model_performance(final, 'F1 Score for Supervised Models')
```

# Semisupervised Learning

semi-supervised learning is a way for computers to learn from a mix of labeled and unlabeled data.


Semi-supervised learning is like having a clever student who uses both the teacher's instructions from the labeled data and the patterns it finds in the unlabeled data to learn. It tries to see if there are similarities or common features between the labeled and unlabeled pictures to figure out things on its own.

By using this combination of labeled and unlabeled data, the computer can become better at understanding new, unseen pictures, even without being explicitly told what they are. This can be really helpful when getting a lot of labeled data is difficult or expensive. The computer can make educated guesses and learn more efficiently by using the additional unlabeled data.

##Self Training Classifier
When just a small quantity of labelled data is available, the Self-Training Classifier is a semi-supervised learning approach that can enhance the performance of a classification model. The approach initially trains the model on labelled data before applying it to unlabeled data to predict labels. These projected labels are added to the training set as "pseudo-labels," which may then be used to iteratively retrain the model.

Self-Training Classifiers are beneficial when getting new labelled data is difficult or expensive. It also integrates easily with current categorization workflows.

Working Conditions: Self-Training The classifier treats the predicted labels as though they are reliable and can increase the model's accuracy.

#Label Propagation
Label Propagation is a semi-supervised algorithm that propagates labels from the labelled data to the unlabeled data using a graph-based methodology. Each data point is represented as a node in the algorithm's created graph, and the edges show how similar the data points are to one another.


Reason for choosing and using: Label Propagation can manage non-linear class borders and is beneficial when there is a smooth transition between classes.


How it operates: Based on their similarity, Label Propagation propagates labels from the labelled data points to the unlabeled data points.


```python
# Train and test the LabelPropagation model on a subset of the final dataset and get its f1 score
score_lp = semi_supervised_model_predictions([model_lp], X_final[12000:27000], y_final[12000:27000], y_final_test[12000:27000])
```

#Label Spreading

Similar to Label Propagation, Label Spreading is a semi-supervised learning system that propagates labels from labelled data to unlabeled data via a diffusion process. Each data point is represented as a node in the algorithm's created graph, and the edges show how similar the data points are to one another.
Motive for choosing: When there are distinct borders between classes and the labelled data points don't fully represent the dataset, label spreading is helpful. Working: Label spreading uses a diffusion mechanism to spread labels from the labelled data points to the unlabeled data points.


```python
# Train and test the LabelSpreading model on a subset of the final dataset and get its f1 score
score_ls = semi_supervised_model_predictions([model_ls], X_final[12000:27000], y_final[12000:27000], y_final_test[12000:27000])
```
